```python
codigos = [
    101, 108, 115, 122, 130, 137, 145, 153, 160, 168,
    175, 183, 190, 198, 205, 213, 220, 228, 235, 243
]


nomes = [
    "Lápis de escrever", "Caneta", "Borracha ", "Apontador",
    "Caderno capa dura", "Caderno espiral", "Régua", "Estojo ",
    "Cola bastão", "Tesoura ", "Marca-texto", "Corretivo ",
    "Bloco de notas", "Agenda ", "Fita adesiva", "Papel A4",
    "Cartolina ", "Envelope pardo", "Grampeador ", "Clips metálico"
]

def busca_binaria(codigos, chave):
    inicio = 0
    fim = len(codigos) - 1
    comparacoes = 0

    while inicio <= fim:
        meio = (inicio + fim) // 2
        comparacoes += 1

        if codigos[meio] == chave:
            return meio, comparacoes
        elif codigos[meio] < chave:
            inicio = meio + 1
        else:
            fim = meio - 1

    return -1, comparacoes


def inserir_ordenado(codigos, nomes, novo_codigo, novo_nome):
    deslocamentos = 0
    i = 0

    while i < len(codigos) and codigos[i] < novo_codigo:
        i += 1
        deslocamentos += 1

    codigos.insert(i, novo_codigo)
    nomes.insert(i, novo_nome)
    return deslocamentos


def menu():
    while True:
        print("\n--- MENU PAPELARIA ---")
        print("1. Buscar produto por código")
        print("2. Inserir novo produto")
        print("3. Listar todos os produtos")
        print("4. Sair")

        opcao = input("Escolha uma opção (1-4): ")

        if opcao == "1":
            try:
                codigo = int(input("Digite o código do produto: "))
                indice, comparacoes = busca_binaria(codigos, codigo)
                if indice != -1:
                    print(f"[✓] Produto encontrado: '{nomes[indice]}' (índice {indice}, {comparacoes} comparações)")
                else:
                    print(f"[✗] Produto não encontrado ({comparacoes} comparações)")
            except ValueError:
                print("Código inválido. Digite apenas números.")

        elif opcao == "2":
            try:
                novo_codigo = int(input("Digite o novo código do produto: "))
                indice, _ = busca_binaria(codigos, novo_codigo)
                if indice != -1:
                    print(f"Esse código já existe para o produto: '{nomes[indice]}'")
                else:
                    novo_nome = input("Digite o nome do novo produto: ")
                    desloc = inserir_ordenado(codigos, nomes, novo_codigo, novo_nome)
                    print(f"Produto '{novo_nome}' inserido com {desloc} deslocamentos.")
            except ValueError:
                print("Código inválido. Digite apenas números.")

        elif opcao == "3":
            print("\n--- LISTA DE PRODUTOS ---")
            for i in range(len(codigos)):
                print(f"{codigos[i]} - {nomes[i]}")

        elif opcao == "4":
            print("Encerrando o programa.")
            break
        else:
            print("Opção inválida. Escolha entre 1 e 4.")


menu()
```
